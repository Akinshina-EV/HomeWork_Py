```python
# Даны две функции:
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# 1. Найти точки пересечения этих функций.
# 2. Построить графики функций в одной системе координат
# 3. Построить график общей функции


from sympy import Symbol, Eq

x = Symbol('x')
f = Symbol('f(x)')
print('Функция 1:')
func_1 = x**3 - 50 * x
Eq(f, func_1)
```

```python
from sympy import Symbol, Eq

g = Symbol('g(x)')
print('Функция 2:')
func_2 = -x**4 + 88 * x**2 - 241
Eq(g, func_2)
```

```python
print('Для вычисления точек пересечения Функции 1 и Функции 2 приравниваем две'
      ' функции друг к другу и решаем уравнение:')
func_3 = Eq(func_1 - func_2, 0)
func_3
```

```python
from sympy import solveset

x1, x2, x3, x4 = solveset(Eq(func_1 - func_2, 0), x)
x1 = x1.evalf(3, chop=True)
y1 = func_1.subs(x, x1)
print(f'Координаты первой точки пересечения Функции 1 и Функции 2: {x1, y1}')
```

```python
x2 = x2.evalf(3, chop=True)
y2 = func_1.subs(x, x2)
print(f'Координаты второй точки пересечения Функции 1 и Функции 2: {x2, y2}')
```

```python
x3 = x3.evalf(3, chop=True)
y3 = func_1.subs(x, x3)
print(f'Координаты третьей точки пересечения Функции 1 и Функции 2: {x3, y3}')
```

```python
x4 = x4.evalf(3, chop=True)
y4 = func_1.subs(x, x4)
print(f'Координаты четвертой точки пересечения Функции 1 и Функции 2: {x4, y4}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10.01, 0.01)
func_1 = x**3 - 50 * x
func_2 = -x**4 + 88 * x**2 - 241
a = plt.plot(x, func_1,  label='f(x) = x^3 - 50x')
b = plt.plot(x, func_2,  label='g(x) = -x^4 + 88x^2 - 241')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
print('График Функции 1 и Функции 2:')
```

```python
x = np.arange(-10, 10.01, 0.01)
func_3 = x**4 + x**3 - 88 * x**2 - 50 * x + 241
c = plt.plot(x, func_3,  label='fg(x) = x^4 + x^3 - 88x^2 - 50x + 241')
plt.xlabel('x')
plt.ylabel('fg(x)')
plt.legend()
plt.grid(True)
print('График общей функции:')
```
